# Task 2 (15 points)

* Download the MNIST dataset and select first 100 images from 1,4,7,8 categories (400 samples total)
* Normalize values to <0,1> and shuffle samples randomly. The np.random.permutation(size) function provides random indexes to shuffle both features and labels synchronously.
* Create and learn the autoencoder model on the dataset. The model will have input layer, three hidden dense layers (30-2-30 neurons) with relu activations and the output layer with sigmoid activations. The second hidden layer will produce two latent features. Learn the model for 300 epoch with the batch_size=5, optimzer Adam (learning_rate=0.008). Use the mean_squared_error as a loss function.
* Generate latent features for the dataset by encoder. Show scatter plot of latent features for each category (1- red, 4- blue, 7-green, 8-magenta)
* For all latent features, generate the corresponding images by the decoder

```python

import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split

(x_train, y_train), (_, _) = mnist.load_data()

selected_categories = [1, 4, 7, 8]
selected_samples = []

for category in selected_categories:
    category_indices = np.where(y_train == category)[0][:100]
    selected_samples.extend(category_indices)

x_selected = x_train[selected_samples]
y_selected = y_train[selected_samples]


x_selected = x_selected.astype('float32') / 255.0

shuffle_indices = np.random.permutation(len(x_selected))
x_selected = x_selected[shuffle_indices]
y_selected = y_selected[shuffle_indices]

x_selected = x_selected.reshape((len(x_selected), -1))

x_train, x_test, _, _ = train_test_split(x_selected, y_selected, test_size=0.2, random_state=42)

input_layer = Input(shape=(784,))
hidden1 = Dense(30, activation='relu')(input_layer)
hidden2 = Dense(2, activation='relu')(hidden1)
hidden3 = Dense(30, activation='relu')(hidden2)
output_layer = Dense(784, activation='sigmoid')(hidden3)

autoencoder = Model(input_layer, output_layer)


autoencoder.compile(optimizer=Adam(learning_rate=0.008), loss='mean_squared_error')


autoencoder.fit(x_train, x_train, epochs=300, batch_size=5, shuffle=True)

encoder = Model(input_layer, hidden2)
latent_features = encoder.predict(x_test)

colors = {1: 'red', 4: 'blue', 7: 'green', 8: 'magenta'}
for category in selected_categories:
    indices = np.where(x_test == category)[0]
    plt.scatter(latent_features[indices, 0], latent_features[indices, 1], c=colors[category], label=str(category))

plt.title('Scatter Plot of Latent Features')
plt.xlabel('Latent Feature 1')
plt.ylabel('Latent Feature 2')
plt.legend()
plt.show()

decoded_images = autoencoder.predict(x_test)


n_images = 5
plt.figure(figsize=(10, 4))

for i in range(n_images):

    plt.subplot(2, n_images, i + 1)
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.title(f'Original\nCategory: {x_test[i]}')
    plt.axis('off')

    plt.subplot(2, n_images, i + n_images + 1)
    plt.imshow(decoded_images[i].reshape(28, 28), cmap='gray')
    plt.title('Generated')
    plt.axis('off')

plt.show()
```

## Task 3 (10 points)

Create a fuzzy system with these parameters
* Two inputs (temperature, water level)
* Single output (heating)
* Temperature range is from 20 to 80 deg. of Celsius
* Level range is from 0 to 200 cm
* Heating range is from 0 to 100 percent
* On the temperature and the water level define the trapezoidal membership functions for low and high
* On heating define the trapezoidal membership functions for low, high and middle
* Calculate the reponse (output) for temperature of 30 deg. and the level of 100 cm
* Show graphs of all fuzzy sets

FUZZY RULES

IF temp is high & ( level is low | level is high) THEN heating is low

IF temp is low and level is low THEN  heating is middle

IF temp is low and level is high THEN heating is high

```python
import matplotlib.pyplot as plt
import numpy as np
import skfuzzy as fuzz

# Define the ranges
temp_range = np.arange(20, 81, 1)
level_range = np.arange(0, 201, 1)
heating_range = np.arange(0, 101, 1)

# Define membership functions
temp_low = fuzz.trapmf(temp_range, [20, 20, 40, 40])
temp_high = fuzz.trapmf(temp_range, [30, 80, 80, 80])

level_low = fuzz.trapmf(level_range, [0, 0, 100, 100])
level_high = fuzz.trapmf(level_range, [80, 200, 200, 200])

heating_low = fuzz.trapmf(heating_range, [0, 0, 30, 30])
heating_middle = fuzz.trapmf(heating_range, [20, 20, 70, 70])
heating_high = fuzz.trapmf(heating_range, [50, 100, 100, 100])

# Plotting the membership functions for temperature, water level, and heating
plt.figure(figsize=(10, 6))

plt.subplot(131)
plt.title('Temperature')
plt.plot(temp_range, temp_low, label='Low')
plt.plot(temp_range, temp_high, label='High')
plt.xlabel('Temperature (C)')
plt.legend()

plt.subplot(132)
plt.title('Water Level')
plt.plot(level_range, level_low, label='Low')
plt.plot(level_range, level_high, label='High')
plt.xlabel('Water Level (cm)')
plt.legend()

plt.subplot(133)
plt.title('Heating')
plt.plot(heating_range, heating_low, label='Low')
plt.plot(heating_range, heating_middle, label='Middle')
plt.plot(heating_range, heating_high, label='High')
plt.xlabel('Heating (%)')
plt.legend()

plt.tight_layout()
plt.show()
```
