# Task 2 (15 points)

* Download the MNIST dataset and select first 100 images from 1,4,7,8 categories (400 samples total)
* Normalize values to <0,1> and shuffle samples randomly. The np.random.permutation(size) function provides random indexes to shuffle both features and labels synchronously.
* Create and learn the autoencoder model on the dataset. The model will have input layer, three hidden dense layers (30-2-30 neurons) with relu activations and the output layer with sigmoid activations. The second hidden layer will produce two latent features. Learn the model for 300 epoch with the batch_size=5, optimzer Adam (learning_rate=0.008). Use the mean_squared_error as a loss function.
* Generate latent features for the dataset by encoder. Show scatter plot of latent features for each category (1- red, 4- blue, 7-green, 8-magenta)
* For all latent features, generate the corresponding images by the decoder

```python
from tensorflow.keras.datasets import mnist
from sklearn.model_selection import train_test_split

#load data
(x_train,y_train),(x_test,y_test)=mnist.load_data()

#subset data
selected_category=[1,4,7,8]
selected_samples=[]
for category in selected_category:
    selected_index=np.where(y_train == category)[0][:100]
    selected_samples.extend(selected_index)

x_train=x_train[selected_samples]
y_train=y_train[selected_samples]

#normalization
x_train=x_train.astype(np.float32)/255.0

#shuffle
shuffled_indices=np.random.permutation(len(x_train))
x_train=x_train[shuffled_indices]
y_train=y_train[shuffled_indices]

#reshape
x_train=x_train.reshape((400,-1))
x_test=x_test.reshape((400,-1))

x_train,x_test,y_train,y_test=train_test_split(x_train, y_train, test_size=0.2, random_state=42)
x_train.shape
```

```python
from tensorflow.keras import models,layers,Sequential,optimizers
import numpy as pd
print(x_train.shape[1])

encoder_input=layers.Input(shape=(x_train.shape[1],))
encoder=Sequential([
    layers.Input(shape=(x_train.shape[1],)),
    layers.Dense(30,activation='relu'),
    layers.Dense(2,activation='relu')
])

decoder_input=layers.Input(shape=(2,))
decoder=Sequential([
    layers.Input(shape=(2,)),
    layers.Dense(30,activation='relu'),
    layers.Dense(x_train.shape[1],activation='sigmoid')
])

autoencoder=Sequential([encoder,decoder])
encoder.summary()
decoder.summary()

autoencoder.compile(optimizer=optimizers.Adam(learning_rate=0.008),loss='mean_squared_error', metrics=['accuracy'])
history=autoencoder.fit( x_train,x_train, batch_size=5, epochs=50, verbose='auto',validation_data=(x_test,x_test))
generated_image=autoencoder.predict(x_test)

```

```python
n_images=5
plt.subplots(figsize=(10,6))
for i in range(n_images):
    plt.subplot(2,n_images,i+1)
    plt.imshow(generated_image[i].reshape(28,28))

    plt.subplot(2,n_images,i+n_images+1)
    plt.imshow(generated_image[i].reshape(28,28))
plt.show()
```

```python
#plot the latent layers onto scatter plot

selected_category=[1,4,7,8]
colors={1:"red", 4:"blue", 7:"green", 8:"magenta"}
latent_features=encoder.predict(x_train)
print(latent_features.shape)
for category in selected_category:
    selected_indices=np.where(y_train==category)[0]
    plt.scatter(latent_features[selected_indices,0],latent_features[selected_indices,1],c=colors[category],label=category)
plt.xlabel('Latent Feature 1')
plt.ylabel('Latent Feature 2')
plt.legend()
plt.show()
    
```

```python
selected_category=[1,4,7,8]
avg_latent_features={}
for category in selected_category:
    selected_index=np.where(y_train==category)[0]
    avg=np.mean(latent_features[selected_index],axis=0)
    avg_latent_features[category]=avg

plt.subplots(figsize=(10,6))
for idx,category in enumerate(selected_category):
    plt.subplot(1,4,idx+1)
    print(avg_latent_features[category].shape)
    gen_image=decoder.predict(np.expand_dims(avg_latent_features[category],axis=0))
    plt.imshow(gen_image.reshape(28,28))

plt.show()

```

## Task 3 (10 points)

Create a fuzzy system with these parameters
* Two inputs (temperature, water level)
* Single output (heating)
* Temperature range is from 20 to 80 deg. of Celsius
* Level range is from 0 to 200 cm
* Heating range is from 0 to 100 percent
* On the temperature and the water level define the trapezoidal membership functions for low and high
* On heating define the trapezoidal membership functions for low, high and middle
* Calculate the reponse (output) for temperature of 30 deg. and the level of 100 cm
* Show graphs of all fuzzy sets

FUZZY RULES

IF temp is high & ( level is low | level is high) THEN heating is low

IF temp is low and level is low THEN  heating is middle

IF temp is low and level is high THEN heating is high

```python
import numpy as np
import skfuzzy as fuzz

temp=np.arange(20,81,1)
water_level=np.arange(0,201,1)
heating=np.arange(0,101,1)

temp_low=fuzz.trapmf(temp,[20, 35, 45, 60])
temp_high=fuzz.trapmf(temp,[40, 55, 65, 80])

water_low=fuzz.trapmf(water_level,[0, 0, 100, 100])
water_high=fuzz.trapmf(water_level,[80, 200, 200, 200])

heating_low=fuzz.trapmf(heating,[0, 0, 20, 30])
heating_middle=fuzz.trapmf(heating,[10, 20, 60, 70])
heating_high=fuzz.trapmf(heating,[50, 100, 100, 100])

plt.figure(figsize=(10,6))

plt.subplot(1,3,1)
plt.plot(temp,temp_low)
plt.plot(temp,temp_high)
plt.xlabel("temperature")

plt.subplot(1,3,2)
plt.plot(water_level,water_low)
plt.plot(water_level,water_high)
plt.xlabel("water level")

plt.subplot(1,3,3)
plt.plot(heating,heating_low)
plt.plot(heating,heating_middle)
plt.plot(heating,heating_high)
plt.xlabel("heating")

plt.show()
```

```python
from skfuzzy import control as ctrl

temp=ctrl.Antecedent(np.arange(20,81,1),'temp')
water_level=ctrl.Antecedent(np.arange(0,201,1),'water_level')
heating=ctrl.Consequent(np.arange(0,101,1),'heating')

temp['low']=temp_low
temp['high']=temp_high

water_level['low']=water_low
water_level['high']=water_high

heating['low']=heating_low
heating['high']=heating_high
heating['middle']=heating_middle

temp.view()
water_level.view()
heating.view()
```

```python
rule1=ctrl.Rule(temp['high'] & (water_level['low'] | water_level['high']), heating['low'])
rule2=ctrl.Rule(temp['low'] & water_level['low'], heating['middle'])
rule3=ctrl.Rule(temp['low'] & water_level['high'] ,heating['high'])

rules=[rule1, rule2, rule3]

heating_ctrl=ctrl.ControlSystem(rules)
sim=ctrl.ControlSystemSimulation(heating_ctrl)
sim.input['temp']=30
sim.input['water_level']=100
sim.compute()

print(sim.output['heating'])
heating.view(sim=sim)
```
