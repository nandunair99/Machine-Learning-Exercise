# Example 3

Draw a graf of the function $y=f(x)=x^2$ and its derivatice $y'=f'(x). Use Tensoflow for derivative calculation.

```python
# Example
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

@tf.function
def f(x):
  y = x*x
  return y

@tf.function
def dfdx(x):
  y = f(x)
  return tf.gradients(y, x)[0]

x = tf.Variable([-3.,-2.,-1.,0.,1.,2.,3.])
print(f(x).numpy())
print(dfdx(x).numpy())

plt.plot(f(x))
plt.plot(dfdx(x))       
```

# Task 2

Draw a graf of the function $y=\sigma(x)=\frac{1}{1+e^{-x}}$ and its derivative $y'=s'(x)$. Use Tensoflow for derivative calculation.

```python
@tf.function
def sgm(x):
    return 1/(1+np.exp(-x))
x=tf.Variable(np.linspace(1,10,10),dtype=tf.float32)
with tf.GradientTape as tape:
    y=sgm(x)
    dy_dx=tape.gradient(y,x)
# Convert tensors to numpy arrays for plotting
x_vals = x.numpy()
y_vals = y.numpy()
dy_dx_vals = dy_dx.numpy()

# Plot the function and its derivative
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="y = f(x) = x^2", color="blue")
plt.plot(x_vals, dy_dx_vals, label="y' = f'(x) = 2x", color="red", linestyle="--")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Function and its Derivative")
plt.legend()
plt.grid()
plt.show()
```
